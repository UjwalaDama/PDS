```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
from plotly.tools import FigureFactory as FF
```

```python
data=pd.read_csv('/content/clean_data.csv')
```

```python
print(data.shape)
```

```python
print(data.columns)
```

```python
data["Frailty"].value_counts()
```

```python
#Performing the 2 sample t-test
t2_test = stats.ttest_ind(data['Weight(lb)'],data['Frailty'])
print('analysis:', t2_test)
```

```python
sns.boxplot(t2_test)
plt.show(t2_test)
```

```python
sns.boxplot(x = data['Frailty'],
            y = data['Grip_strength(kg)'])
plt.show(t2_test)
```

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(data, hue="Frailty", height = 3);
plt.show()
#Pairwise Scatter plot:Pair plot
#Observation:Here if we observe weight vs grip strength person with weight more than 135 and grip strength
```

```python
sns.FacetGrid(data, hue="Frailty", height =5) \
.map(sns.histplot,"Age") \
.add_legend();
plt.show();
# Classification of Frailty based on Age
```
